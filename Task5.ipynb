```python
        #Social Media – Post Engagement Analyzer
def analyze(reactions_list):
  total_likes=0
  for i in reactions_list:
    total_likes+=i

  post_status = ""
  if total_likes >=1000:
    post_status = "The post is popular"
  else:
    post_status = "Normal engagement"
  return total_likes, post_status

post_reactions = [450, 300, 300]
total_likes, post_status = analyze(post_reactions)
print(f"Total Likes: {total_likes}")
print(f"Post Status: {post_status}")
```

```python
        # Healthcare – Medicine Stock Alert System
def medicine(med_name, current_stock):

  status = ""
  if current_stock < 10:
    status = "low stock"
  else:
    status = "stock sufficient"

  return {
    "medicine": med_name,
    "stock": current_stock,
    "status": status
  }

medicine_data = medicine("Amoxicillin", 6)
print(f"Medicine Stock: {medicine_data['stock']}")
print(f"Status: {medicine_data['status']}")
```

```python
          # Agriculture – Rainfall Adequacy Checker
def rainfall_checker(rainfall_data, required_level):
    total = 0
    for rain in rainfall_data:
        total = total + rain

    average = total / len(rainfall_data)

    print("Average Rainfall:", round(average))


    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")

season_rainfall = [65, 70, 80, 75, 68, 74]
required = 70

rainfall_checker(season_rainfall, required)
```

```python
        #Social Media – Duplicate Account Detection
def duplicate(username):
  unique_user=set()
  duplicate_found=False
  for user in username:
    if user in unique_user:
      duplicate_found=True
      break
    else:
      unique_user.add(user)


  if duplicate_found:
      print("Duplicate Accounts Found: Yes")
  else:
      print("Duplicate Accounts Found: No")

user_list = ["rutu123", "amit45", "sneha01", "rutu123", "karan77"]
duplicate(user_list)
```

```python
          #Application Health Checker
def health(error_count):

    print("Error Count:", error_count)

    if error_count == 0:
        print("System Status: Healthy")
    elif error_count <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


health(7)
```

```python
#Student Attendance Eligibility System
def eligibility(attendance_list):

    total_classes = len(attendance_list)
    present_count = 0

    for status in attendance_list:
        if status == "P":
            present_count += 1

    percentage = (present_count / total_classes) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


attendance = ["P", "P", "A", "P", "P"]
eligibility(attendance)
```

```python
#Smart Electricity Bill Analyzer
def calculate_bill(units):

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage


total_bill, usage_status = calculate_bill(250)

print("Total Bill:", total_bill)
print("Usage Status:", usage_status)
```
