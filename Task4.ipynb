```python
# Mobile Recharge Validation System

def recharge():

    valid_plans = [199, 299, 399, 599]

    print("MObile Recharge")

    while True:
        try:
            amount = int(input("\nEnter recharge amount (Min ₹50): "))

            if amount < 50:
                print(f"Error: {amount} is too low.")

            elif amount not in valid_plans:
                print("Invalid Plan! .")

            else:
                print(f"Processing... ₹{amount} recharge successful!")
                break

        except ValueError:
            print("Invalid input. Please enter numbers only.")

        retry = input("Try another amount? (y/n): ").lower()
        if retry != 'y':
            print("Exiting system. Goodbye!")
            break

if __name__ == "__main__":
    recharge()
```

```python
      #Inventory Reorder Alert System

def check_reorder(stock_data):
    for product in stock_data:
        qty = stock_data[product]

        if qty < 15:
            print(product, "- Reorder Alert")
        else:
            print(product, "- Stock OK")



inventory = {
    "Rice": 10,
    "Wheat": 25,
    "Sugar": 14,
    "Oil": 30,
    "Salt": 8
}

check_reorder(inventory)
```

```python
            # Student Result Processing System
def check_result(marks):
    total = 0

    for m in marks:
        total = total + m

    avg = total / len(marks)
    if len(marks) == 0:
        return "No marks entered"

    if avg >= 50:
        return "Pass"
    else:
        return "Fail"



marks = input("Enter marks separated by space: ")
marks_list= list(map(int, marks.split()))

result = check_result(marks_list)

print("Result:", result)
```

```python
      # Cab Fare Estimator with Retry Option

def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (distance * 12)

    if peak == "yes":
        fare = fare + (fare * 0.25)

    return round(fare, 2)


while True:
    km = float(input("Enter distance in km: "))

    if km < 0:
        print("Distance cannot be negative")
        continue

    peak_hour = input("Is it peak hour? (yes/no): ").lower()

    if peak_hour not in ["yes", "no"]:
        print("Please enter yes or no")
        continue

    total_fare = calculate_fare(km, peak_hour)

    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ").lower()

    if retry != "yes":
        break
```

```python
# Employee Attendance Eligibility
def check_eligibility(att_list):
    if not att_list:
        return "No attendance entered"

    present_count = 0
    for day in att_list:
        if day == "p":
            present_count += 1

    percentage = (present_count / len(att_list)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

data = input("Enter attendance (P/A separated by space):")
attendance = data.split()

result = check_eligibility(attendance)
print("Attendance: {}%".format((attendance.count("p") / len(attendance) * 100) if attendance else 0))
print("Status:", result)
```

```python
# Password Strength Checker
def password(ped):
  if len(ped) < 8:
    return "Weak Password"
  digit = False
  special = False
  special_char = "@#$"

  for ch in ped:
    if ch.isdigit():
      digit = True
    if ch in special_char:
      special = True

  if digit and special:
    return "Strong password"
  else:
    return "Weak password"

password_input = input("Enter password :")
res = password(password_input)
print("Password status :", res)
```
