```python
# Problem 1: Smart Parking Lot Management System
def parking(capacity, logs):
    parked = 0
    peak = 0
    for entry in logs:
        if entry == "IN":
            parked += 1
        elif entry == "OUT":
            if parked > 0:
                parked -= 1
                if parked > peak:
                    peak = parked
    print("Currently Parked Vehicles:", parked)
    if parked > capacity:
        print("Parking Status: Full (Alert!)")
    else:
        print("Parking Status: Available")

parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
# delivery time calculator
def delivery(distance, traffic, weather):

    time = distance * 5
    if traffic == "High":
        time = time + 15
    elif traffic == "Medium":
        time = time + 10
    else:
        time = time + 5

    if weather == "Rainy":
        time = time + 10
    elif weather == "Stormy":
        time = time + 20

    print("Estimated Delivery Time:", time, "minutes")

delivery(8, "High", "Rainy")
```

```python
# Movie Theatre Seat Occupancy Analyzer
def analyze_occupancy(total_seats, booked_list):
    booked = len(booked_list)
    percentage = (booked / total_seats) * 100

    print("Occupancy:", int(percentage), "%")
    if percentage == 100:
        print("Show Status: Housefull")
    elif percentage >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")

analyze_occupancy(200, [1]*150)
```

```python
# Cloud Server Load Classification System
def classify_load(cpu_readings):
    total = 0
    for value in cpu_readings:
        total = total + value
        average = total / len(cpu_readings)
    print("Average CPU Load:", int(average), "%")

    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

classify_load([45, 60, 70, 85, 90])
```

```python
# Smart Classroom Resource Usage Monitor
def resources(resource_dict):
    overused = []

    for resource in resource_dict:
        hours = resource_dict[resource]
        if hours > 8:
            overused.append(resource)
            if len(overused) > 0:
              print("Overused Resources:", ", ".join(overused))
              print("Energy Alert: Yes")
        else:
              print("Overused Resources: None")
              print("Energy Alert: No")

resources({"Projector": 6, "AC": 9, "Lights": 4})
```

```python
# Online Event Registration Capacity Controller
def manage_registrations(capacity, registrations):

    if registrations > capacity:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlist = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)

manage_registrations(100, 105)
```
