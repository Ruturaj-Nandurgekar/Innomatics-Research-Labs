```python
    # USER LOGIN CHECK

username = "admin"   #correct username and pass
password = "1234"

username_input= input("username: ")   #input form user
password_input= input("password: ")

if username_input== username and password_input== password:
  print("Login Successful")        #condition to check
else:                              #input in if or else
  print("Invalid Credentials")
```

```python
      # PASS/FAIL ANALYSE

marks = [45, 78, 90, 33, 60]    # stdudent marks

pass_count = 0   #initialize the value to 0
fail_count = 0

for mark in marks:   #loop to iterate
    if mark >= 50:          #check >= 50
        pass_count += 1     #yes then increment else inc the fail
    else:
        fail_count += 1

print("total number of pass students: ", pass_count)
print("total number of fail students: ", fail_count)
```

```python
        # Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]    #original list

cleaned=[]    #empty list to store

for n in names:     #to iterate on it
    clean_ver=n.strip().lower()      #remove space and lower case
    cleaned.append(clean_ver)

print("Cleaned list: ",cleaned)    #printing the res
```

```python
# Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]

for m in messages:
    size = len(m)
    print("message:", m)
    print("length is:", size)

    if size > 10:
        print("This message is greater than 10 characters")

    print()
```

```python
# Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_total = 0

for l in logs:
    if l == "ERROR":
        error_total += 1

print("total number of error are: ", error_total)
```
