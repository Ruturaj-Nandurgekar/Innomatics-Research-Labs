```python
          # Smart Light Controller

class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.state = "OFF"

    def set_state(self, action):
        self.state = action

    def show(self):
        print(self.light_name + " is " + self.state)

b = SmartLight("Bedroom Light")
b.set_state("ON")
b.show()
```

```python
          # Employee ID Card System

class EmpCard:
    def __init__(self, name, eid, dept):
        self.name = name
        self.eid = eid
        self.dept = dept

    def print_card(self):
        print("Employee ID Card")
        print("Name: " + self.name)
        print("ID: " + self.eid)
        print("Department: " + self.dept)

e1 = EmpCard("Rahul", "EMP102", "IT")
e1.print_card()
```

```python
        # Mobile Contact Record

class MobileContact:
    def __init__(self, cname, ph):
        self.cname = cname
        self.ph = ph

    def save(self):
        print("Contact Saved")
        print("Name: " + self.cname)
        print("Phone: " + self.ph)

c1 = MobileContact("Anita", "9876543210")
c1.save()
```

```python
      # Product Price Tag Generator

class PriceTag:
    def __init__(self, pname, cost):
        self.pname = pname
        self.cost = cost

    def display(self):
        print("Product: " + self.pname)
        print("Price: ₹" + str(self.cost))

p1 = PriceTag("Headphones", 2499)
p1.display()
```

```python
      # Movie Rating Display System

class MovieInfo:
    def __init__(self, title, stars):
        self.title = title
        self.stars = stars

    def show(self):
        print("Movie: " + self.title)
        print("Rating: " + str(self.stars) + " / 5")

m1 = MovieInfo("Inception", 4.8)
m1.show()
```

```python
          # Delivery Address Manager

class DeliveryInfo:
    def __init__(self, cust, addr):
        self.cust = cust
        self.addr = addr

    def disp_details(self):
        print("Delivery Details")
        print("Customer: " + self.cust)
        print("Address: " + self.addr)

d1 = DeliveryInfo("Suman", "Hyderabad")
d1.disp_details()
```
