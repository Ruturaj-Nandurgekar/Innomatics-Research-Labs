```python
        # Unique Words in a Sentence
text=input("enter sentence: ")      #for input
text1=text.split()                  #split into list
unique=set(text1)                   #remove duplicate
tunique=len(unique)                 #get count

print("Unique words count: ",tunique)       #print
print("Unique words:",unique)
```

```python
        #Highest Salary from Employee Data
employees = {
    "Ravi": 75000,              #dictionary
    "Anita": 68000,
    "Kiran": 72000
}

top_earn=""     #var
highest=0
for name in employees:
  current_sal=employees[name]           #compare

  if current_sal>highest:               #checking  highest
    highest=current_sal
    top_earn=name

    print("Highest salary :",top_earn,"-",highest)

```

```python
  #Find Maximum and Minimum Values
numbers = [45, 22, 89, 10, 66]


max_val = numbers[0]
min_val = numbers[0]

for n in numbers:           #loop through
   if n > max_val:
     max_val = n             #check max
   if n < min_val:
     min_val = n             #check min


print("List :",numbers)     #print
print("Max :",max_val)
print("Min :",min_val)
```

```python
      #Count Products Above a Price Threshold
prices = [450, 1200, 899, 1500, 300]

count=0       #init count

for i in prices:        #loop
  if i > 1000:           #check
     count+=1

print("Product above 1000 :", count)        #print
```

```python
        #Calculate Attendance Percentage
attendance = ["P", "P", "A", "P", "P"]

total=len(attendance)

count=0

for i in attendance:
    if i=="P":
       count+=1

Percentage=(count/total)*100
print("Attendance percentage :", Percentage)
```

```python
        # Count Character Frequency
word = "pythonp"
freq={}

for char in word:
   if char in freq:
      freq[char]+=1
   else:
      freq[char]=1

print(freq)
```
