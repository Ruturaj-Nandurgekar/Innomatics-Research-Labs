```python
        # 1 Employee Performance Bonus Eligibility
employees = {
  "Ravi": 92,
  "Anita": 88,
  "Kiran": 92,
  "Suresh": 85
}
maxscore=max(employees.values())
winn=[]

for name in employees:
  if employees[name]==maxscore:
     winn.append(name)

name=",".join(winn)
print(f"Top Performers Eligible for Bonus: {name} (Score: {maxscore})")
```

```python
        #2  Search Query Keyword Analysis
query="Buy mobile phone buy phone online"   #input
clean=query.lower().replace(".", "")       #tolowert
words=clean.split()     #to split
count={}        # dict to store
for i in words:
  if i in count:
    count[i]=count[i]+1     # if exist +1
  else:
    count[i]=1
print("keyword frequency analysis :")     #display
for i,counts in count.items():
  print(f"{i}:{counts}")
```

```python
          #3 Sensor Data Validation
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
readings=[]     #to store
for hour in range(len(sensor_readings)):
  read=sensor_readings[hour]

  if read %2==0:
    readings.append((hour,read))
print("valid sensor readings(Hour, Value):")
print(readings)
```

```python
        #4 Email Domain Usage Analysis
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
counts={}       #dict to store count
for email in emails:
    parts=email.split("@")        #split and take 2nd part
    domain=parts[1]

    if domain in counts:          #count update in dict
     counts[domain]=counts[domain]+1
    else:
      counts[domain]=1

total_email=len(emails)

print("email domain usage analysis :")

for domain_key in counts:           #display
  count=counts[domain_key]
  percentage=(count/total_email)*100

  print(f"{domain_key}:{percentage}%")
```

```python
          #5 Sales Spike Detection
sales = [1200, 1500, 900, 2200, 1400, 3000]

total_sales=sum(sales)        #avg sales
avg_sales=total_sales/len(sales)

threshold=avg_sales*1.30      # more thwan 30% cal

for index, amount in enumerate(sales):        #detect spike days
    if amount > threshold:
        day_number = index + 1
        print(f"Day {day_number}: {amount}")
```
